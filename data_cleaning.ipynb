## **Data Preparation**

**Membuat Kolom 'Lokasi' Berdasarkan Kolom 'Latitude' dan 'Longitude'**

```python
# Inisialisasi geocoder
geolocator = Nominatim(user_agent="geoapi")

# Menggunakan RateLimiter agar tidak diblokir oleh server
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)

def get_location(lat, lon):
    try:
        location = reverse((lat, lon), exactly_one=True)
        if location and "address" in location.raw:
            address = location.raw['address']
            return address.get('city') or address.get('town') or address.get('village') or address.get('county')
        return None
    except:
        return None

# Buat kolom 'Lokasi'
data_wisata['Lokasi'] = data_wisata.apply(lambda row: get_location(row['Latitude'], row['Longitude']), axis=1)
data_restoran['Lokasi'] = data_restoran.apply(lambda row: get_location(row['Latitude'], row['Longitude']), axis=1)
data_hotel['Lokasi'] = data_hotel.apply(lambda row: get_location(row['Latitude'], row['Longitude']), axis=1)
```

**Menghapus Kolom 'Latitude' dan 'Longitude'**

```python
data_wisata = data_wisata.drop(['Latitude', 'Longitude'], axis=1)
```

```python
data_restoran = data_restoran.drop(['Latitude', 'Longitude'], axis=1)
```

```python
data_hotel = data_hotel.drop(['Categories', 'Latitude', 'Longitude'], axis=1)
```
